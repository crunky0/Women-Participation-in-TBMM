```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
import matplotlib.pyplot as plt

data = pd.read_csv("city info/formatted/turkiye.csv")

# Model Using Last 8 Terms
data = data[-8:]

# X for Feature Y for Label
X = data[['Year']]
y = data['Female MP Count']

poly_features = PolynomialFeatures(degree=2)
X_poly = poly_features.fit_transform(X)

model = LinearRegression().fit(X_poly, y)

future_years = [2023, 2028, 2033, 2038]
future_X = pd.DataFrame({'Year': future_years})
future_X_poly = poly_features.transform(future_X)
future_y_pred = model.predict(future_X_poly)
y_pred = model.predict(X_poly)

for i in range(len(future_years)):
    print(future_years[i], round(future_y_pred[i]))

mse = mean_squared_error(y, y_pred)
r2 = r2_score(y, y_pred)
print('Mean Squared Error:', mse)
print('R-squared:', r2)

plt.scatter(data['Year'], data['Female MP Count'], color='blue', label='Actual Data')

# Sort the data for smoother visualization
sorted_indices = np.argsort(data['Year'])
sorted_year = data['Year'].values[sorted_indices]
sorted_y_pred = model.predict(poly_features.transform(pd.DataFrame({'Year': sorted_year})))

plt.plot(sorted_year, sorted_y_pred, color='red', linewidth=2, label='Regression Line')
plt.scatter(future_years, future_y_pred, color='green', marker='o', label='Future Prediction')
y_min = min(min(y), min(future_y_pred))
y_max = max(max(y), max(future_y_pred))

plt.ylim(y_min - 5, y_max + 5)
plt.plot(future_years, future_y_pred, color='green', linewidth=2, linestyle='dashed')
plt.xlabel('Year')
plt.ylabel('Female MP Count')
plt.title('Polynomial Regression')
plt.legend()
plt.show()
```
